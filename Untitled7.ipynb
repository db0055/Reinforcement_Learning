```python
import numpy as np
```

```python
x = np.random.randint(0, 1000, 2000)
```

```python
y = x + np.random.normal(0, 1000, 2000)
```

```python
np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x, y)
plt.show()
```

```python
y = 100 - x + np.random.normal(0, 1000, 2000)

np.corrcoef(x, y)
```

```python
import pandas as pd

df = pd.DataFrame({'a': np.random.randint(0, 1000, 2000)})
df['b'] = df['a'] + np.random.normal(0, 1000, 2000) # positively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.normal(0, 1000, 2000) # negatively correlated with 'a'
df['d'] = np.random.randint(0, 1000, 2000) # not correlated with 'a'

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(6, 6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
